```python
#libraries
import pandas as pd
import seaborn as sns
import scipy.stats as st
```

```python
#Load dataset
df = pd.read_csv("Baseball.csv")
df.head()
```

```python
#information about the dataset
df.info()
```

```python
#Mean of runs scored
df.RS.mean()
```

```python
# mean of run scored by arizona team (ARI)
df.loc[df.Team == "ARI"].RS.mean()
```

```python
#mean of runs scored by the arizona team since 2005
df.loc[(df.Team == "ARI") & (df.Year >= 2005)].RS.mean()
```

```python
# what is the average of runs by the chicago team (CHC)
# Average before 2000 and after 2005
```

```python
#average of runs by the chicago team (CHC)
df.loc[df.Team == "CHC"].RA.mean()

```

```python
#Average before 2000
df.loc[(df.Team == "CHC") & (df.Year < 2000)].RA.mean()

```

```python
#Average AFTER 2005
df.loc[(df.Team == "CHC") & (df.Year > 2005)].RA.mean()
```

```python
df.loc[(df.Team == "CHC") & ((df.Year < 2000) | (df.Year > 2005))].RA.mean()
```

```python
#mean and median
print(df.W.mean())
print(df.W.median())
```

```python
#what is the mean of Wins of the baltimore team (BAL) until 2000
df.loc[(df.Team == "BAL") & (df.Year <= 2000)].W.median()
```

```python
#Mode
print(df.OBP.mean())
print(df.OBP.median())
print(df.OBP.mode())
df.OBP.hist()
```

```python
# What is the mode of OBP during 2010
df.loc[df.Year == 2010].OBP.mode()
```

```python
# What is the most commom season ranking for BOS
df.loc[df.Team == "BOS"].RankSeason.mode()
```

```python
# Standar deviation and variance

#OOBP

print(f"Mean: {df.OOBP.mean()}")
print(f"STD sample: {df.OOBP.std()}")
print(f"STD population: {df.OOBP.std(ddof = 0)}")
print(f"Variance: {df.OOBP.var()}")
```

```python
#what is the standard deviation of the sample of BA
# and the variance of the population of BA

print(f"STD sample: {df.BA.std()}")
print(f"Variance: {df.BA.var(ddof = 0)}")
df.BA.hist()
```

```python
# Function for descriptive statistics

df.describe()
```

```python
#coeficient of variation for runs scored variable

cv = df.RS.std() / df.RS.mean()
print(cv)
```

```python
#coeficient of variation for Wins

cv = df.W.std() / df.W.mean()
print(cv)
```

```python
#covariance with continuous variable
df_co = df[["RS", "RA", "W"]]
print(df_co.cov())
df_co.cov()
```

```python
# pick three of the variables and applie covariance

df_co = df[["BA", "G", "W"]]
df_co.cov()
```

```python
# correlation
df_co.corr()
```

```python
#visualization heat map
sns.heatmap(data = df_co.corr(),
            annot=True,
            center = 0,
            cmap = 'coolwarm',
            linewidths=1,
            linecolor='black')
```

#Conclusions

###From this data, the following conclusions can be drawn:

Team Performance Trends:

Arizona (ARI) consistently scores an average number of runs higher after 2005.
Chicago (CHC) allowed fewer runs before 2000 compared to after 2005.
Statistical Summaries:

Wins (W) show a higher mean than median, indicating potential outliers or skewed distribution.
The Baltimore (BAL) team's win count until 2000 highlights median wins, emphasizing stability.
On-Base Percentage (OBP):

OBP follows a non-normal distribution with its mode differing by year (e.g., 2010).
Variability:

High coefficients of variation in runs scored (RS) and wins indicate significant variability in team performance.
Relationships:

Covariance and correlation analysis reveal positive relationships between runs scored (RS), runs allowed (RA), and wins (W).
Heatmap visualization highlights the strength and direction of these correlations.
These insights provide an understanding of team performance, variability, and key relationships across the dataset.
